```python
#Line Plotting

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4])
y = 2*x
y1 = np.array([3,5,7,9])
plt.plot(x,y, linestyle ='dashed', color='red', marker = 'o', linewidth = 2.5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Graph')

plt.fill_between(x,y,y1, color = 'blue', alpha = 0.3, label = 'Filled Area')

plt.legend()
plt.show()
```

```python
#Bar Plotting

#Vertical Bar
import matplotlib.pyplot as plt
import numpy as np
fruits = ['apple', 'banana', 'mango', 'orange']
sales = [400, 350, 300, 450]
plt.bar(fruits, sales, color = 'green', width=0.2)
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Fruits Sale')
plt.show()
```

```python
#Horizontal Bar
import matplotlib.pyplot as plt
import numpy as np
fruits = ['apple', 'banana', 'mango', 'orange']
sales = [400, 350, 300, 450]
plt.barh(fruits, sales, color = 'maroon', height=0.8)
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Fruits Sale')
plt.show()
```

```python
#Scatter Plotting

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([160, 165, 170, 175, 180, 185, 190, 195, 200, 205])
y1 = np.array([55, 58, 60, 62, 64, 66, 68, 70, 72, 74])
bubble_sizes = [30, 80 ,150, 200, 270, 310, 380, 415, 476, 500]


x2 = np.array([160, 165, 170, 175, 180, 185, 190, 195, 200, 205])
y2 = np.array([50, 52, 54, 56, 58, 64, 66, 68, 70, 72])

plt.scatter(x1, y1, s = bubble_sizes, alpha = 0.5, linewidth = 1.5, edgecolor = 'black', color = 'violet', label = 'Group 1')
plt.scatter(x2, y2, color = 'darkcyan', label = 'Group 2')

plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Comparison of Height and Weight between two groups')

plt.legend()
plt.show()
```

```python
#Histogram Plotting

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randn(1000)
plt.hist(x, bins = 30, color = 'skyblue', edgecolor = 'black')
plt.xlabel('Values')
plt.ylabel('Frequencies')
plt.title('Histogram')
plt.show()
```

```python
#Pie Chart Plotting

import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabel = np.array(['Apples', 'Bananas', 'Mangoes', 'Cherries'])
myexplode = [0.2, 0.0, 0.0, 0.0]
mycolor = ['lightgreen', 'skyblue', 'pink', 'yellow']
plt.pie(y, labels = mylabel, startangle=90, explode = myexplode, colors = mycolor)
plt.legend(loc = 'upper right', title = '4 Fruits')
plt.show()
```
